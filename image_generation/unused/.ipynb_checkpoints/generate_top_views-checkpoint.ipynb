```python
# Find the min and max on image indices to perform the crop
def cropbg(cv_img):
    # Find pixels that are not zero
    cv_img_clone = cv_img.copy()
    np_img = np.asarray(cv_img_clone)
    print(np_img.shape)
    ind_fg = np.where(np_img[:, :, 0]!=[0])
    i_min = ind_fg[0].min()
    i_max = ind_fg[0].max()
    j_min = ind_fg[1].min()
    j_max = ind_fg[1].max()
    # for i in range(len(ind_fg[0])):
    #     print(ind_fg[0][i], ind_fg[1][i])
    disp_img = cv.rectangle(cv_img_clone, (j_min, i_min), (j_max, i_max), (255,0,0), 2)
    crop_img = cv_img[i_min:i_max, j_min:j_max]
    pyplot.imshow(disp_img)
    pyplot.show()
    pyplot.imshow(crop_img)
    pyplot.show()

    return crop_img
    # print(ind_fg.size)
```

```python
# Make the background black
# count = 10
for file in blender_pngs:
    img = cv.imread(file)
    cv_img = cv.imread(file)
    img_np = np.asarray(img)
    img_np[img_np <= 64] = 0
    img_np[img_np > 64] = 255
    pil_img_new = Image.fromarray(img_np.astype(np.uint8))
    cv_img_new = np.array(pil_img_new)
    cv_img_new = cv_img_new[:, :, ::-1].copy()

    disp_img = cv.hconcat([cv_img, cv_img_new])
    # cv.imshow(disp_img)
    pyplot.imshow(disp_img)
    pyplot.show()

    crop_img = cropbg(cv_img_new)
    # find_top_surface(crop_img)
    file_name = file.split("/")[2]
    print(file_name)
    out_file_path = top_view_pngs_path+file_name
    print(out_file_path)
    cv.imwrite(out_file_path, crop_img)
    # count = count -1
    # if count < 0:
    #     break
```
